## Data Cleaning

```python
 df = df.dropna(subset=["rating_count"])
```

```python
df
```

```python
df.isnull().sum()
```

```python
df["discounted_price"] = (
    df["discounted_price"]
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
)

df["discounted_price"] = pd.to_numeric(df["discounted_price"])
```

```python
df["discounted_price"].dtype
```

```python
df["actual_price"] = (
    df["actual_price"]
    .str.replace("₹","",regex=False)
    .str.replace(",","",regex=False)
)

df["actual_price"] = pd.to_numeric(df["actual_price"])
# df["actual_price"] = (
#     df["actual_price"]
#     .str.replace("₹", "", regex=False)
#     .str.replace(",", "", regex=False)
# )

# df["actual_price"] = pd.to_numeric(df["actual_price"])
```

```python
df["discount_percentage"]= (
    df["discount_percentage"] 
    .str.replace("%","", regex=False)
)
df["discount_percentage"] = pd.to_numeric(df["discount_percentage"])
```

```python
avg_rating = df.groupby('main_category')[ 'rating'].mean().sort_values(ascending = False)
```

## Exploratory Data Analysis (EDA)

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(avg_rating.index , avg_rating.values)
plt.title('avg rating by main category')
plt.xlabel('main category')
plt.ylabel('avg rating')
plt.xticks(rotation = 45)
plt.show()
```

```python
df.groupby('main_category')['discount_percentage'].mean().sort_values(ascending=False)
```

```python
plt.scatter(df['discount_percentage'], df['rating'])
plt.title("discount percentage vs rating")
plt.xlabel("discount percentage")
plt.ylabel("rating")
plt.show()
```

```python
df['discount_percentage'].corr(df['rating'])
```

```python
df.groupby('main_category')[['actual_price','discounted_price','rating']].mean()
```

```python
df.to_csv("amazon_cleaned.csv", index = False)
```

## Key Insights
- Electronics contains the highest number of products.
- Home Improvement has the highest average discount.
- Office Products have the highest average rating.
- Discounts show only a weak relationship with ratings.
